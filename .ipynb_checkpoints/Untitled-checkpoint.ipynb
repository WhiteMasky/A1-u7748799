```python
# (a) Read image and resize
image = cv2.imread('image1.jpg')
image = cv2.resize(image, (384, 256))
```

```python
# (b) Separate RGB channels and display
b, g, r = cv2.split(image)
cv2.imshow('Blue Channel', b)
cv2.imshow('Green Channel', g)
cv2.imshow('Red Channel', r)
```

```python
# (c) Compute and display histograms
def display_histogram(image, title):
   plt.figure()
   plt.title(title)
   plt.hist(image.ravel(), 256, [0, 256])
   plt.show()

display_histogram(b, 'Blue Channel Histogram')
display_histogram(g, 'Green Channel Histogram')
display_histogram(r, 'Red Channel Histogram')
```

```python
# (d) Histogram equalization
def histogram_equalization(image):
   hist, bins = np.histogram(image.ravel(), 256, [0, 256])
   cdf = hist.cumsum()
   cdf_normalized = cdf * hist.max() / cdf.max()
   cdf_m = np.ma.masked_equal(cdf, 0)
   cdf_m = (cdf_m - cdf_m.min()) * 255 / (cdf_m.max() - cdf_m.min())
   cdf = np.ma.filled(cdf_m, 0).astype('uint8')
   img_new = cdf[image]
   return img_new

b_eq = histogram_equalization(b)
g_eq = histogram_equalization(g)
r_eq = histogram_equalization(r)

display_histogram(b_eq, 'Equalized Blue Channel Histogram')
display_histogram(g_eq, 'Equalized Green Channel Histogram')
display_histogram(r_eq, 'Equalized Red Channel Histogram')

image_eq = cv2.merge((b_eq, g_eq, r_eq))
cv2.imshow('Equalized Image', image_eq)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
