#### Problem 1

1.1  
__How many nodes are reachable in t steps from the central node?__
<br>
Given a Cayley tree, from the central node, we can reach (exclude the central node)
$$N_k=
  k(k - 1)^{t-1},\quad t \leq P 
$$
nodes in k steps. Since all nodes are reachable within a value less than or equal to $P$ in a Cayley tree, $t \leq P$ is added here; if $t$ is greater than $P$, then it is clear that it is able to reach all nodes. 
<br>
The equation could be 
$$N_k=
  k(k - 1)^{t-1}+1,\quad t \leq P 
$$
including the central node.

1.2  
__What is the degree distribution of the network?__
<br>
We can derive that the total number of nodes in the Cayley tree is $$N = 1+\sum_{i=1}^P (k-1)^{i-1}k$$ (plus 1 node means the center)
<br>
The degree distribution is:
<br> 
If $P=0$, then there will be only one node have no degree.
<br>
When $P\not=0$
- $Degree_1 = (k-1)^{P-1}k$ nodes have degree 1 (the leaves)
- $Degree_k = N - Degree_1$ nodes have degree k

1.3  
__Calculate the diameter $d_{max}$ of a (k, P) Cayley tree__
<br>
The diameter is the maximal distance between any two leaves nodes connected in the network. In Cayley tree, the distance of any leaf node from the central node is P, we can calculate that $$d_{max}=2P$$
The k value does not affect its diameter.

1.4  
__Express the diameter $d_{max}$ in terms of the total number of nodes N__
<br>
We can further extrapolate the equation in 1.2 
$$
\begin{align}
  N &= 1+\sum_{i=1}^P (k-1)^{i-1}k \\
   &=  1 + k((k-1)^P-1)\\
\end{align}
$$
By the upper equation we can derive the expression of $d_{max}$ as a function of N.
$$
\begin{array}{r}
N=1+k\left[(k-1)^{d_{max} / 2}-1\right] \\
\left(\frac{N-1}{k}\right)=(k-1)^{d_{max} / 2}-1 \\
\left(\frac{N-1}{k}\right)+1=(k-1)^{d_{max} / 2} \\
\frac{d_{max}}{2} \ln (k-1)=\ln \left[1+\frac{N-1}{k}\right] \\
d_{max}=\frac{2}{\ln (k-1)} \ln \left[1+\frac{N-1}{k}\right] . 
\end{array}
$$

1.5  
__Does the (k, P) Cayley tree display the small-world property, why or why not__
<br>
Yes, the network of Cayley tree displays the small-world distance property. Specifically, a small-world network is defined to be a network where the typical distance $L$ between two randomly chosen nodes (the number of steps required) grows proportionally to the logarithm of the number of nodes $N$ in the network (Watts & Strogatz 1998), that is:
$$
L \propto log N
$$
while the global clustering coefficient is not small.According to the equation in 1.4, when $ N \gg 1 $, $$d_{max}≃\frac{2}{\ln (k-1)}  [\frac{\ln{N}}{\ln{k}}]$$ wich complies the concept of small world network.

2.2  
__Argue that there is at least some nodes with less than < k > connections in this small component.__  
The emergence of the giant component is $\left\langle k \right\rangle  = 1 $, so in this problem, graph $G(N,p)$ has $\left\langle k \right\rangle=(N-1)p>1$.  
If we remove all the nodes involved in the giant connected component from the entire graph, resulting in a graph $G'$, then $G'$ with those nodes only in small components can be considered as a graph that has not yet emerged a giant connected component, so it is in a subcritical state or on the critical point. At this point, the corresponding $\left\langle k' \right\rangle \leq 1 < \left\langle k \right\rangle $.  
Therefore, given a small component consisting of $s$ nodes that is not part of the giant component, there is at least some nodes with less than $\left\langle k \right\rangle$ connections in this small component.  

__Consider adding one new edge to such a node, what is the probability that such a new edge is within the small component, or to the giant component__  

Consider the following situation (since $G(N,p)$ is an undirected graph, there is no difference bwtween start and end): 
- Start from nodes in this small component, end at any nodes:$${s \choose 1}  {N-1 \choose 1} = s(N-1)$$
- Start from nodes in this small component, end in the small component, that is within the small component:$${s \choose 2} = \frac{s(s-1)}{2}$$
- Start from nodes in this small component, end in the giant component, that is to the giant component: $${s \choose 1}  {N-s-1 \choose 1} = s(N-s-1)$$  

Hence, we got
$$ Probability\; that\; such\; a\; new\; edge\; is\; within\; the\; small\; component = \frac{s-1}{2(N-1)}$$  
$$ Probability\; that\; such\; a\; new\; edge\; is\; to\; the\; giant\; component = \frac{N-s-1}{N-1}$$

3.1  

For an arbitrary node $u$, it is connected to its four nearest neighbors in the lattice (three neighbors in the case of nodes on the boundary, two nodes in the case of nodes on the corner according to the figure below).

\begin{align*}
\sum_{v \in N, v \neq u} d(u, v)^{-2} &\leq \sum_{j=1}^{2 n-2}(4 j)\left(j^{-2}\right) \\
&= 4 \sum_{j=1}^{2 n-2} j^{-1} \\
&\leq 4+4 \ln (2 n-2) \leq 4 \ln (6 n) .
\end{align*}
$$


3.2  
In this step we are going to estimate the total number of steps spent in phase $j$, we need to estimate the probability of being a long-range contact from the message holder
Suppose we are now in phase $j$, if we want to end in this step, we want the message to enter the set $B_{j}$ of nodes within lattice distance $2^{j}$ of $t$.  
There are at least
$$
1+\sum_{i=1}^{2^{j}} i=\frac{1}{2} 2^{2 j}+\frac{1}{2} 2^{j}+1>2^{2 j-1}
$$
nodes in $B_{j}$, each is within lattice distance $2^{j+1}+2^{j}<2^{j+2}$ of $u$.  
From 3.1, for an arbitrary node $u$ has one directed edge to another node $v$, that is one node has a single long-range contact $v$, which is $d(u, v)^{-2} / \sum_{v \neq u} d(u, v)^{-2}$, so that the probability that $v$ is chosen is at least $\left[4 \ln (6 n) d(u, v)^{2}\right]^{-1}$. Thus, each node in $B_{j}$ has a probability of at least $\left(4 \ln (6 n) 2^{2 j+4}\right)^{-1}$ of being the long-range contact of $u$.  
If any of these nodes serves as the long-range contact of $u$, it will become $u$'s closest neighbor to $t$. As a result, the message is likely to enter $B_{j}$ with a probability of at least
$$
\frac{2^{2 j-1}}{4 \ln (6 n) 2^{2 j+4}}=\frac{1}{128 \ln (6 n)}
$$
The current message holder is $u$, and target is $t$. In the 2-dimensional case, according to the definiton of the decentralized algorithm $A$, A is in phase 0 when the lattice distance to t is at most 2. Thus, the initial value of j is at most log n.  In phase $j$ we got $\log (\log n) \leq j<\log n$.  
Let the random variable $X_j$ denote the total number of steps spent in phase $j$, we have
$$
\begin{aligned}
E [X_{j}] & =\sum_{i=1}^{\infty} \operatorname{Pr}\left[X_{j} \geq i\right] \\
& \leq \sum_{i=1}^{\infty}\left(1-\frac{1}{128 \ln (6 n)}\right)^{i-1} \\
& =128 \ln (6 n) .
\end{aligned}
$$
Analogically, for $j = log\;n$, $E [X_{j}] \leq 128 \ln (6 n)$ also works. For $0 \leq j \leq \log (\log n)$, the algorithm can spend at most $\log n$ steps in phase $j$ even if all nodes pass the message to a local contact, $E [X_{j}] \leq 128 \ln (6 n)$ still holds.


3.3  
As for time complexity, let $X$ denotes the total number of steps spent by the algorithm, we have
$$
X=\sum_{j=0}^{\log n} X_{j}
$$
and so by linearity of expectation we have $E [X] \leq(1+\log n)(128 \ln (6 n)) \leq \alpha (\log n)^{2}$ for a suitable choice of an constant $\alpha$. In conclusion, through the decentralized algorithm $A$, the expected delivery time is at most $\alpha (\log n)^{2}$, that is, the time complexity of algorithm $A$ is $O (\log n)^{2}$, so all message could be delivered within $O (\log n)^{2}$ steps in expectation.

```python
import networkx as nx

# Assuming G1 is your original network
n = G1.number_of_nodes()  # Number of nodes
k = int(sum(dict(G1.degree()).values()) / n)  # Average degree

# Set rewiring probability
p = 0.1  # This is a common choice, but you can adjust it as needed

# Generate Watts-Strogatz network
G_ws = nx.watts_strogatz_graph(n, k, p)

# Now G_ws is a Watts-Strogatz network with the same number of nodes and average degree as G2
nx.draw(G_ws, with_labels=False, node_size=3, node_color='g', edge_color='y',width = 0.1)
plt.show()
```

```python
# Assuming G1, G2, G3, G4, G5 are your networks
networks = [G1, G2, G_er1, G_er2, G_ws]
network_names = ['G1', 'G2', 'G_er1', 'G_er2', 'G_wg']

data = {}

for name, G in zip(network_names, networks):
    n = G.number_of_nodes()
    m = G.number_of_edges()
    avg_degree = 2*m/n
    avg_clustering = nx.average_clustering(G)
    transitivity = nx.transitivity(G)
    degree_sequence = [d for n, d in G.degree()]
    max_degree = max(degree_sequence)
    min_degree = min(degree_sequence)
    # Get the largest connected component
    largest_cc = max(nx.connected_components(G), key=len)
    subgraph = G.subgraph(largest_cc)
    # Calculate diameter and radius on the subgraph
    diameter = nx.diameter(subgraph)
    radius = nx.radius(subgraph)
    # diameter = nx.diameter(G)
    # radius = nx.radius(G)
    closeness_centrality = nx.closeness_centrality(G)
    max_closeness_centrality = max(closeness_centrality.values())
    betweenness_centrality = nx.betweenness_centrality(G)
    max_betweenness_centrality = max(betweenness_centrality.values())

    data[name] = [n, m, avg_degree, avg_clustering, transitivity, max_degree, min_degree, diameter, radius, max_closeness_centrality, max_betweenness_centrality]

df = pd.DataFrame(data, index=['Number of nodes', 'Number of edges', 'Average degree', 'Average clustering', 'Transitivity', 'Max degree', 'Min degree', 'Diameter', 'Radius', 'Max closeness centrality', 'Max betweenness centrality'])

print(df)
```
